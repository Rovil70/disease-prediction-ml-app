```python
import pandas as pd

df = pd.read_csv("dataset.csv")

df.head()
```

```python
df["symptoms"] = df["symptoms"].str.lower()
```

```python
all_symptoms = set()

for s in df["symptoms"]:
    for item in s.split(","):
        all_symptoms.add(item.strip())

all_symptoms = list(all_symptoms)

len(all_symptoms), all_symptoms[:10]
```

```python
for sym in all_symptoms:
    df[sym] = df["symptoms"].apply(lambda x: 1 if sym in x else 0)
```

```python
df.drop("symptoms", axis=1, inplace=True)

df.head()
```

```python
df.to_csv("processed_dataset.csv", index=False)
```
